```python

# gather dataset start
import pandas as pd
import requests
import time
```

```python
national_dex = 1, 2, 3, ..., 493

# api route, https://pokeapi.co/api/v2/pokemon/{national_dex}

def generation_from_dex(national_dex):
    if 1 <= national_dex <= 151:
        return 1
    elif 152 <= national_dex <= 251:
        return 2
    elif 252 <= national_dex <= 386:
        return 3
    elif 387 <= national_dex <= 493:
        return 4
    else:
        return None


def get_pokemon_row(national_dex):
    url = f"https://pokeapi.co/api/v2/pokemon/{national_dex}"
    response = requests.get(url)
    response.raise_for_status()

    data = response.json()

    pokemon = data["name"]

    types = sorted(data["types"], key=lambda x: x["slot"])
    type_1 = types[0]["type"]["name"]
    type_2 = types[1]["type"]["name"] if len(types) > 1 else None

    stats = {
        stat["stat"]["name"]: stat["base_stat"]
        for stat in data["stats"]
    }

    hp = stats["hp"]
    atk = stats["attack"]
    defense = stats["defense"]
    spa = stats["special-attack"]
    spd = stats["special-defense"]
    spe = stats["speed"]

    total = hp + atk + defense + spa + spd + spe

    return {
        "national_dex": national_dex,
        "pokemon": pokemon,
        "type_1": type_1,
        "type_2": type_2,
        "HP": hp,
        "Atk": atk,
        "Def": defense,
        "SpA": spa,
        "SpD": spd,
        "Spe": spe,
        "Total": total,
        "generation": generation_from_dex(national_dex),
    }

```

```python
rows = []

for national_dex in range(1, 494):
    row = get_pokemon_row(national_dex)
    rows.append(row)

    # optional: be polite to the API
    time.sleep(0.05)

pokemon_df = pd.DataFrame(rows)

pokemon_df.head()
```

```python
# gather dataset end
pokemon_df.to_csv("./sample_data/pokemon_gen1_to_gen4.csv", index=False)
```
